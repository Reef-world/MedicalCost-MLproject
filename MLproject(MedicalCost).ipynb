# **medical cost prediction**
*our model will try and predict the medical cost of patient deepending on factors like age, do they smoke, how many childern do they have,and the region they reside(in the united states)..*

# **preprocessing the data**
in this step we will import the libraries, read the dataset and get a view of what we're working on.. then start preparing the data before training

```python
import pandas as pd
import numpy as np
```

```python
MedicalCost = pd.read_csv('/content/MedicalCost.csv')
MedicalCost
```

```python
MedicalCost.info()
```

**like shown form the data information, since there's no null vlaues then we will start right away with the encoding for the catagorical attributes such as sex, smoker and region**

```python
MedicalCost['sex'].value_counts()
```

```python
MedicalCost['smoker'].value_counts()
```

```python
MedicalCost['region'].value_counts()
```

```python
sex_encoded = pd.get_dummies(MedicalCost['sex']) #function to encode values
sex_encoded
```

```python
sex_encoded= sex_encoded.drop('female', axis=1)
```

```python
sex_encoded
```

```python
MedicalCost = pd.concat([MedicalCost, sex_encoded], axis=1)
```

```python
MedicalCost
```

```python
MedicalCost = MedicalCost.drop('sex', axis=1)
```

```python
MedicalCost = MedicalCost.rename(columns={'male': 'sex'})
```

```python
MedicalCost.head()
```

**as shown the column sex now is encoded in a way that 0 indecate the value female and the 1 indecate the value male**

**now we will repeat the same process with the features smoker and region**

```python
smoker_encoded = pd.get_dummies(MedicalCost['smoker'])
smoker_encoded
```

```python
smoker_encoded= smoker_encoded.drop('no', axis=1)
```

```python
MedicalCost = pd.concat([MedicalCost, smoker_encoded], axis=1)
```

```python
MedicalCost = MedicalCost.drop('smoker', axis=1)
```

```python
MedicalCost = MedicalCost.rename(columns={'yes': 'smoker'})
```

```python
MedicalCost.head()
```

```python
MedicalCost['region'].value_counts()
```

**as shown, the attribute region holds 4 different vslues, so unlike the the two past attributes which had binary values, but in this case we will use the binary encoder emthod and merge the values into one column that holds binary values**

```python
region_encoded = pd.get_dummies(MedicalCost['region'])
region_encoded
```

```python
MedicalCost = pd.concat([MedicalCost, region_encoded], axis=1)
```

**the binary encoded columns was added as seperate columns so now we need to merge them and rename them with the attribute name they represent**

```python
MedicalCost['region'] = MedicalCost[['northeast', 'northwest', 'southeast', 'southwest']].apply(lambda x: ''.join(map(str, x)), axis=1)
```

```python
MedicalCost.drop(['northeast', 'northwest', 'southeast', 'southwest'], axis=1, inplace=True)
```

```python
MedicalCost.head()
```

```python
MedicalCost['region'].value_counts()
```

**0010 indecate northeast**

**1000 indecate northwest**

**0001 indecate southeast**

**0100 southwest**

**the last step is to change the column type into int**

```python
MedicalCost['region'] = MedicalCost['region'].astype(int)
```

```python
MedicalCost.info()
```

**now that all columns represented as numerical type, we can start the training process**
